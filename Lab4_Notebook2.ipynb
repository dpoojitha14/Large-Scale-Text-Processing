Reading output of hadoop WordCount.

```r
text = readLines("part-r-00000")

data = read.table(file="part-r-00000", sep="\t", quote="", comment.char="")
```

```r
wordcloud(words = data$V1, freq = data$V2, min.freq = 1,max.words=200, random.order=FALSE, rot.per=0.35, colors=brewer.pal(8, "Dark2"))
```
